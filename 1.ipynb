```python
import matplotlib.pyplot as plt

# Sample data for the number of crops and predictions
number_of_crops = [1, 2, 3, 4, 5]
low_predictions = [10, 15, 20, 25, 30]
high_predictions = [30, 40, 50, 60, 70]

# Create the first graph for low predictions
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(number_of_crops, low_predictions, marker='o', color='b', label='Low Prediction')
plt.xlabel('Number of Crops')
plt.ylabel('Predictions')
plt.title('Low Prediction Scenario')
plt.legend()

# Create the second graph for high predictions
plt.subplot(1, 2, 2)
plt.plot(number_of_crops, high_predictions, marker='s', color='g', label='High Prediction')
plt.xlabel('Number of Crops')
plt.ylabel('Predictions')
plt.title('High Prediction Scenario')
plt.legend()

plt.tight_layout()
plt.show()
```
